```python
# Calculate the total number of students
total_students = len(df)

# Calculate the number of students who meet each criterion(Yes = 0)
enrolled_count =(df['full_enrollment'] == 0).sum()
placement_count = (df['placement'] == 0).sum()
completed_program_count = (df['completion_status'] == 0).sum()

# Calculate percentages
enrolled_percentage = (enrolled_count / total_students) * 100
placement_percentage = (placement_count / total_students) * 100
completed_program_percentage = (completed_program_count / total_students) * 100

# Print the results
print(f"Percentage of students fully enrolled: {enrolled_percentage:.2f}%")
print(f"Percentage of students with a job after graduation: {placement_percentage:.2f}%")
print(f"Percentage of students who completed the program: {completed_program_percentage:.2f}%")
```

```python
from scipy.stats import chi2_contingency

# Function to perform chi-square test
def chi_square_test(variable):
    contingency_table = pd.crosstab(df[variable], df['completion_status'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    return p

# List of categorical columns (all binary columns except a few)
categorical_columns = [
'full_enrollment', 'placement', 'gap12_mnthscore', 'newfood_score',
        'foodserv_24mnth', 'Age at intake', 'food_stamps', 'cash_assistance',
        'credential', 'edu_level_encoded'
]

# Perform Chi-Square test for each variable
for col in categorical_columns:
    p_value = chi_square_test(col)
    print(f"Chi-Square Test for {col}: p-value = {p_value}")

p_values = {col: chi_square_test(col) for col in categorical_columns}
# Convert to lists for visualization
test_names = list(p_values.keys())
p_values_list = list(p_values.values())

# Create visualization
plt.figure(figsize=(10, 5))
plt.barh(test_names, p_values_list, color="skyblue")
plt.axvline(x=0.05, color="red", linestyle="--", label="Significance Threshold (0.05)")
plt.xlabel("p-value")
plt.ylabel("Chi-Square Test")
plt.title("Chi-Square Test Results")
plt.legend()
plt.xscale("log")  # Log scale for better visualization of small values
plt.show()
```

```python
from scipy.stats import f_oneway

# List of score-based variables
score_columns = [ 'q24_reasons', 'q25_goals', 'q26_career','q27_food']

# Filtering data based on completion status
complete = df[df['completion_status'] == 1]
incomplete = df[df['completion_status'] == 0]

# Perform ANOVA for each variable
anova_results = {}
for col in score_columns:
    f_stat, p_value = f_oneway(complete[col], incomplete[col])
    anova_results[col] = (f_stat, p_value)
    print(f"ANOVA Test for {col}: F-statistic = {f_stat}, p-value = {p_value}")

```

```python
import matplotlib.pyplot as plt

# Extract variables and values for visualization
variables = list(anova_results.keys())
f_stats = [anova_results[var][0] for var in variables]
p_values = [anova_results[var][1] for var in variables]

# Create a dual-axis bar plot
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot F-statistics
ax1.barh(variables, f_stats, color='skyblue', label="F-statistic")
ax1.set_xlabel("F-statistic")
ax1.set_ylabel("Variables")

# Secondary axis for p-values
ax2 = ax1.twiny()
ax2.scatter(p_values, variables, color='red', label="p-value")
ax2.set_xlabel("p-value (red dots)")

# Add a significance threshold line
ax2.axvline(x=0.05, color="black", linestyle="--", label="Significance Threshold (0.05)")

# Title and legend
plt.title("ANOVA Test Results")
ax1.legend(loc="lower right")
ax2.legend(loc="upper right")
plt.xscale("log")  # Log scale for better visualization of small values

plt.show()
```

```python
from scipy.stats import linregress
import matplotlib.pyplot as plt

# List of score-based variables
score_columns = ['q24_reasons', 'q25_goals', 'q26_career','q27_food']

# Perform Linear Regression for each variable
regression_results = {}
for col in score_columns:
    slope, intercept, r_value, p_value, std_err = linregress(df[col], df['completion_status'])
    regression_results[col] = (slope, p_value, r_value**2)
    print(f"Linear Regression for {col}: p-value = {p_value}, R-squared = {r_value**2}")

# Visualization
plt.figure(figsize=(10, 5))
variables, p_values = zip(*[(k, v[1]) for k, v in regression_results.items()])
plt.barh(variables, p_values, color="lightcoral")
plt.axvline(x=0.05, color="red", linestyle="--", label="Significance Threshold (0.05)")
plt.xlabel("p-value")
plt.ylabel("Variable")
plt.title("Linear Regression P-values for Score-Based Variables")
plt.legend()
plt.xscale("log")  # Log scale for better visualization
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Select predictor and target variables
X = df[['q24_reasons', 'q25_goals', 'q26_career', 'q27_food']]
y = df['completion_status']

# Apply Polynomial Transformation (Degree 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Split data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Train Polynomial Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Model Evaluation
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

# Print results
print(f"Train R-squared: {train_score:.3f}")
print(f"Test R-squared: {test_score:.3f}")

# Visualize predictions vs. actual values
plt.scatter(y_test, model.predict(X_test), color='blue', alpha=0.5)
plt.xlabel("Actual Completion Status")
plt.ylabel("Predicted Completion Status")
plt.title("Polynomial Regression Predictions vs. Actual")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.show()
```
