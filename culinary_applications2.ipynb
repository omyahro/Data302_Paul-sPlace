```python
import pandas as pd
import numpy as np

culinary = pd.read_excel("C:/Users/eppst/OneDrive/Documents/Howard/HU - DATA302/Pauls Place Documents/Travon/Application process notes.xlsx", usecols=[4,10,17])
culinary
```

```python
culinary = culinary.fillna(0)
culinary
```

```python
import pandas
from sklearn import linear_model


X = culinary[['# applicants who Passed Math & Reading', 'Goal # spots to fill']]
y = culinary['Applications Submitted']

regr = linear_model.LinearRegression()
regr.fit(X.values, y.values)

# predict the # of Applications Submitted where # of applicants who Passed Math and Reading is 26 and the Goal # spots to fill is 12
predict_applications = regr.predict([[26,12]])
print(predict_applications)
```

```python
def reg_predict():
    for i in range(10,20):  # passed math/reading
        for x in range(10,16):  # goal spots to fill
            print(f"{x}: goal spots to fill, {i}: passed math/reading, for {regr.predict([[i,x]])}: applications submitted")
```

```python
reg_predict()
```

```python
# redefined the function because an applicant has to pass the math/reading portion.  The number of applicants who passed math/reading
# can't be less than the number of spots to fill.
def reg_predict():
    for i in range(10,20+1):  # passed math/reading
        for x in range(i,i+1):  # goal spots to fill
            print(f"{x}: goal spots to fill, {i}: passed math/reading, for {regr.predict([[i,x]])}: applications submitted")
```
